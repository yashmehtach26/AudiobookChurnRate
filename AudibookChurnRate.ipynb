```python
shuffled_indices = np.arange(inputs.shape[0])
np.random.shuffle(shuffled_indices)
shuffled_inputs = scaled_inputs[shuffled_indices]
shuffled_outputs = output[shuffled_indices]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(shuffled_inputs, shuffled_outputs, test_size=0.1, random_state=365)
```

```python
validation_size = int(x_train.shape[0]/10)
x_validation = x_train[:validation_size]
x_train = x_train[validation_size:x_train.shape[0]]

y_validation = y_train[:validation_size]
y_train = y_train[validation_size:y_train.shape[0]]
```

```python
print (x_train.shape, y_train.shape)
print (x_test.shape, y_test.shape)
print (x_validation.shape, y_validation.shape)
```

```python
input_size = 10
depth = 50
output_size = 2
batch_size = 100
max_epochs = 100
early_stopping = tf.keras.callbacks.EarlyStopping(patience=2)
model = tf.keras.Sequential([
    tf.keras.layers.Dense(depth, activation='relu'),
    tf.keras.layers.Dense(depth, activation='relu'),
    tf.keras.layers.Dense(output_size, activation='softmax')
])
model.compile(optimizer='Adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(x_train,
            y_train,
            epochs=max_epochs,
            batch_size=batch_size,
            validation_data=(x_validation, y_validation),
            verbose=2,
             callbacks = [early_stopping])
```
